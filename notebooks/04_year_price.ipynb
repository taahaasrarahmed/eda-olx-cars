```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# 1) Load & prep
df = pd.read_csv('../OLX_cars_dataset00.csv')
```

```python
# Basic scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(
    df[df['Year'].map(df['Year'].value_counts()) >= 2]['Year'], 
    df[df['Year'].map(df['Year'].value_counts()) > 2]['Price'] / 100000, 
    alpha=0.5
)
plt.xlabel('Model Year')
plt.ylabel('Price (lac PKR)')
plt.title('Price vs. Model Year')
plt.grid(True)
plt.show()
```

```python
# Check by make (top 3)
top_makes = df['Make'].value_counts().head(3).index
for brand in top_makes:
    brand_df = df[df['Make'] == brand]
    plt.figure(figsize=(8, 6))
    plt.scatter(
        brand_df['Year'],
        brand_df['Price'] / 100000,  # price in lacs
        alpha=0.5
    )
    plt.title(f"{brand} - Price vs Model Year")
    plt.xlabel("Model Year")
    plt.ylabel("Price (lac PKR)")
    plt.grid(True)
    plt.show()
```

```python
# Check by model (top 3)
top_models = df['Model'].value_counts().head(3).index
for model in top_models:
    model_df = df[df['Model'] == model]
    plt.figure(figsize=(8, 6))
    plt.scatter(
        model_df['Year'],
        model_df['Price'] / 100000,  # price in lacs
        alpha=0.5
    )
    plt.title(f"{model} - Price vs Model Year")
    plt.xlabel("Model Year")
    plt.ylabel("Price (lac PKR)")
    plt.grid(True)
    plt.show()
```
